```python
def make_column(city,column_name,extra_index1 = 'default',extra_index2 = 'default',extra_index3 = 'default'):
        city = city.replace(' ','%20')
        request = "https://api.waqi.info/feed/{}/?token={}".format(city, apikey)
        response = requests.get(request).json()
        try:
            if extra_index3 != 'default':
                return response['data'][column_name][extra_index1][extra_index2][extra_index3]
            elif extra_index2 != 'default':
                return response['data'][column_name][extra_index1][extra_index2]
            elif extra_index1 != 'default':
                return response['data'][column_name][extra_index1]
            else: 
                return response['data'][column_name]
        except Exception as e:
            return None
        
```
