```python
#set up for webscraping
from bs4 import BeautifulSoup
import requests
import re
```

```python
#set up for webscraping
url = 'https://data.mongabay.com/cities_pop_01.htm'

r = requests.get(url)

soup = BeautifulSoup(r.text, 'html.parser')
```

```python
#start webscraping to get a list of the 1000 most populated cities in the world
city_name_list = []

find_table = soup.find('table', class_="boldtable")
```

```python
#start webscraping to get a list of the 1000 most populated cities in the world
find_row = find_table.find_all('tr')
```

```python
#start webscraping to get a list of the 1000 most populated cities in the world
for row in find_row:
    city_find = row.find('td')
    city_text = city_find.get_text()
    
    cleaned_city1 = city_text.split(',')[0]
    cleaned_city2 = cleaned_city1.split('(')[0]
    city_cleaned = cleaned_city2.lower().title()
    print(city_cleaned)
    
    city_name_list.append(city_cleaned)
```

```python
one_thousand_largest_cities = city_name_list[2:]
```

```python
#create waqi data frame with cities column
waqi_df = pd.DataFrame({'cities':one_thousand_largest_cities})
```

```python
#get_waqi_data takes in a string as input and returns a dictionary as output.
#get_waqi_data creates a dictionary of all the responses from the air quality index api for each city
def get_waqi_data(city):
    city = city.replace(' ','%20')
    request = "https://api.waqi.info/feed/{}/?token={}".format(city, apikey)
    response = requests.get(request).json()
    try:
        print(city)
        return response
    except Exception as e:
        #print(e)
        return "something wrong with request {}".format(e)
```

```python
#collect the responses for each city in a dictionary
waqi_data_dict = {}
```

```python
for city in waqi_df['cities']:
    waqi_data_dict[city] = get_waqi_data(city)
```

```python
#testing the output
waqi_data_dict['Tokyo']['data']
```

```python
#add_aqi takes in a string as input and returns an integer as output.
def add_aqi(name):
    try:
        aqi = waqi_data_dict[name]['data']['aqi']
        return aqi
    except Exception as e:
        return None
```

```python
#add_latitude takes in a string as input and returns a float as output.
def add_latitude(name):
    try:
        latitude = waqi_data_dict[name]['data']['city']['geo'][0]
        return latitude
    except Exception as e:
        return None
```

```python
#add_longitude takes in a string as input and returns a float as output.
def add_longitude(name):
    try:
        latitude = waqi_data_dict[name]['data']['city']['geo'][1]
        return latitude
    except Exception as e:
        return None
```

```python
#add_pm25 takes in a string as input and returns a float as output.
def add_pm25(name):
    try:
        latitude = waqi_data_dict[name]['data']['iaqi']['pm25']['v']
        return latitude
    except Exception as e:
        return None
```

```python
#add_pm10 takes in a string as input and returns a float as output.
def add_pm10(name):
    try:
        latitude = waqi_data_dict[name]['data']['forecast']['daily']['pm10'][0]['avg']
        return latitude
    except Exception as e:
        return None
```

```python
#add_o3 takes in a string as input and returns a float as output.
def add_o3(name):
    try:
        latitude = waqi_data_dict[name]['data']['forecast']['daily']['o3'][0]['avg']
        return latitude
    except Exception as e:
        return None
```

```python
#add_uvi takes in a string as input and returns a float as output.
def add_uvi(name):
    try:
        latitude = waqi_data_dict[name]['data']['forecast']['daily']['uvi'][0]['avg']
        return latitude
    except Exception as e:
        return None
```

```python
#add_date takes in a string as input and returns a string as output.
def add_date(name):
    return "11-30-2021"
```

```python
#Add the remainig columns to the waqi data frame
waqi_df['aqi'] = waqi_df['cities'].apply(add_aqi)
waqi_df['latitude'] = waqi_df['cities'].apply(add_latitude)
waqi_df['longitude'] = waqi_df['cities'].apply(add_longitude)
waqi_df['pm25'] = waqi_df['cities'].apply(add_pm25)
waqi_df['pm10'] = waqi_df['cities'].apply(add_pm10)
waqi_df['o3'] = waqi_df['cities'].apply(add_o3)
waqi_df['uvi'] = waqi_df['cities'].apply(add_uvi)
waqi_df['date'] = waqi_df['cities'].apply(add_date)
```
