```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = 'https://data.mongabay.com/cities_pop_01.htm'

r = requests.get(url)

soup = BeautifulSoup(r.text, 'html.parser')
```

```python
city_name_list = []

find_table = soup.find('table', class_="boldtable")
```

```python
find_row = find_table.find_all('tr')
```

```python
for row in find_row:
    city_find = row.find('td')
    city_text = city_find.get_text()
    
    cleaned_city1 = city_text.split(',')[0]
    cleaned_city2 = cleaned_city1.split('(')[0]
    city_cleaned = cleaned_city2.lower()
    print(city_cleaned)
    
    city_name_list.append(city_cleaned)
```
